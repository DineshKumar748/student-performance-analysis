```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load data
df = pd.read_csv("/Users/dineshkumarmuthusamy/Desktop/python/Loop/students.csv")
# Display first few rows of the dataframe
print(df.head())
```

```python
# Average per student
df["average"] = df[["math", "science", "english"]].mean(axis=1)
# Display names and their average scores
print(df[["name", "average"]])
```

```python
# Pass / fail
passed = 0
failed = 0
for avg in df["average"]:
    if avg >= 60:
        passed += 1
    else:
        failed += 1
print(f"\nNumber of students who passed: {passed}")
print(f"Number of students who failed: {failed}")
```

```python
# Subject averages
subject_avg = df[["math", "science", "english"]].mean()
print("\nSubject Averages:")
print(subject_avg)
```

```python
# Best student
top_student = df.loc[df["average"].idxmax()]

print("Passed students:", passed)
print("Top student:", top_student["name"])
```

```python
# Plot
subject_avg.plot(kind="bar", title="Average Scores by Subject")
plt.ylabel("Score")
plt.show()
```

```python
# Plot
# X and Y values
x = df["math"]
y = df["science"]

# Scatter plot
plt.scatter(x, y)

# Create best-fit line
m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x + b)

# Labels
plt.xlabel("Math Score")
plt.ylabel("Science Score")
plt.title("Math vs Science Scores (with Trend Line)")

plt.show()
```
